```python
india=country_df("India")
```

```python
def line_plot(c_df):
    x = c_df["Date_reported"]
    y = c_df["Cumulative_cases"]
    plt.plot(x, y)
    plt.show()
```

```python
line_plot(india)
```

```python
def line_plot_m(c_name):
    c_df = country_df(c_name)
    x = c_df["Date_reported"]
    y = c_df["Cumulative_cases"]
    plt.plot(x, y)
    plt.show()
```

```python
line_plot_m("Iceland")
```

```python
uk = country_df("The United Kingdom")
```

```python
germany = country_df("Germany")
```

```python
plt.plot(uk["Date_reported"],uk["Cumulative_cases"], 
         label = "UK", color = "Green")
plt.plot(germany["Date_reported"],germany["Cumulative_cases"], 
         label = "Germany", color = "yellow")
plt.legend()
plt.xlabel("Date")
plt.ylabel("Cum Cases")
plt.title("Comparision of UK and Germany")
plt.show()
```
